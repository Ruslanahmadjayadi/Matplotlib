```python
import matplotlib.pyplot as plt # mengimport matplotlib menjadi plt 
import numpy as np # mengimport numpy menjadi np
```

```python
#membuat sample data
x = [1, 2, 3, 4]
y = [1, 4, 4, 3]

fig, ax = plt.subplots() 
# membuat figure dan axes dengan method ini dan menambahkan 2 var untuk menampung nya
# fig untuk figure dan ax untuk axes
ax.plot(x, y)
# .plot ini untuk melakukan plotting pada axes dan akan menvisualkan data terhadap x dan y
```

```python
#membuat sample data
x = [1, 2, 3, 4]
y = [1, 4, 3, 3]

plt.plot(x, y)# plt.plot sama seperti yang sebelumnya tetapi cara lebih simple
```

```python
# figure merupakan media dasar untuk melakukan visualisasi data dengan melakukan matplotlib
# axes merupakan area gambarnya atau area untuk melakukan plotting
# axis mengindikasikan sumbu seperti sumbu x dan sumbu y
```

```python
fig = plt.figure() # untuk membuat figure tanpa adanya axes
```

```python
fig, ax=plt.subplots() # untuk membuat sebuah figure dengan adanya axes
```

```python
fig, axs = plt.subplots(2, 3) 
# untuk membuat figure dengan beberapa axes seperti contoh yang terdapat parameter 2,3 tersebut 
#untuk menghasilkan figure dengan axes dengan plotting 2 baris dan 3 colom
```

```python
#ooStyle
x = np.linspace(0, 2, 100)
x

# method dataset ini untuk mengenerate sejumlah deret bilangan yang ditentukan oleh ketiga parameter
# yaitu dengan 100 buah dengan ukuran angka dari bilangan 0 sampe bilangan  2
```

```python
fig, ax = plt.subplots() #membuat sebuah figure dan axes

# membuat plotting tiga jenis data pada data axes dengan cara oos style
ax.plot(x, x, label='linear')
# untuk menghasilkan garis dan menggambarkan garis lurus karena nilai x dan y selalu sama
ax.plot(x, x**2, label='quadratic') 
# untuk mengahasilkan garis dan menggambarkan garis sedikit naik dikarenakan x dan y nya 
#tidak sama karena y nya kuadrat (**2)
ax.plot(x, x**3, label='cubic')
# untuk mengahasilkan garis dan menggambarkan garis lebih naik dikarenakan x dan y nya 
#tidak sama karena y nya kubik (**3)

ax.set_xlabel('xlabel') # method set ini untuk memberi nilai pada nilai sumbu x
ax.set_ylabel('ylabel') # method set ini untuk memberi nilai pada nilai sumbu y
ax.set_title('Simple plot') # untuk memberi tittle pada axes
ax.legend() # untuk membuat sebuah legend
```
