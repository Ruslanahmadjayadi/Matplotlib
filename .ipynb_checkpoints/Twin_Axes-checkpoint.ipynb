# Twin Axes
- Mempelajari Twin Axes dengan Matplotlib

## Import Modules
- Mempelajari cara membuat Violin Plot dengan Matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Kasus Twin Axes 1
- Pemanggilan `np.linspace` ini akan membutuhkan 3 parameter
- Parameter 1 (1) untuk menentukan nilai terkecil
- Parameter 2 (10) untuk menentukan nilai terbesar
- Parameter 3 (25) untuk menentukan data point yang akan dihasilkan

```python
x = np.linspace(1, 10, 25)
x
```

- Plotting pertama akan melakukan plotting nilai x dengan menggunakan fungsi exponensial (biru dan marker kotak)
- `twinx()` untuk menghasilkan axes baru dan ditempatkan diatas axes pertama dan ditampung pada ax2
- Plotting Kedua yaitu plotting logaritmi (Merah dan marker bulat)
- Plotting dilakukan dengan cara oostyle

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='lower left')

plt.show()
```

## Kasus Twin Axes 2
- `x = np.random.randint(16, 40, size=30)` akan menghasilkan bilangan integer random yang diatur dengan parameternya
- Parameter 1 (16) untuk menentukan nilai terkecil
- Parameter 2 (40) untuk menentukan nilai terbesar
- Parameter 3 (size=30) untuk menentukan data point yang akan dihasilkan

```python
x = np.random.randint(16, 40, size=30)
x
```

- Untuk melakukan plotting memerlukan suatu fungsi yang akan mengkonversi celcius to farenheit

```python
def C2F(celsius=0):
    return (celsius * 1.8) + 32

def konversi_sumbu(ax1): 
    y1, y2 = ax1.get_ylim() 
    ax2.set_ylim(C2F(y1), C2F(y2)) 
    ax2.figure.canvas.draw() 

fig, ax1 = plt.subplots() 
ax2 = ax1.twinx() 

ax1.callbacks.connect('ylim_changed', konversi_sumbu) 
ax1.plot(x) 

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius') 
ax2.set_ylabel('Fahrenheit') 

fig.suptitle('Temperatur Udara') 
plt.show()
```
