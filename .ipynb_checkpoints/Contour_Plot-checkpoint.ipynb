# Contour Plotting
- Mempelajari cara membuat Contour Plotting dengan Matplotlib

## Import Modules

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Membentuk Sample Dataset
- Membentuk suatu deret bilangan yang menghasilkan nilai dari -3 sampai 3 dengan jumlah 100 bilangan
- `np.meshgrid(n, n)` untuk membentuk data surface (data area) yang disimpan dalam variabel x untuk data pertama dan y untuk data kedua
- Variabel Z akan menampung hasil akar kuadrat dari `np.sqrt(x**2 + y**2)`

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## Contour Plot
- Contour Plot yaitu salah satu cara untuk memvisualkan data dengan format 3 dimensi menjadi format 3 dimensi

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Filled Contour Plot
- Untuk memenuhi area contour plot dengan warna

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
