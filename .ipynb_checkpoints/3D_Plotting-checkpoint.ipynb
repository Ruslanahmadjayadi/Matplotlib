# 3D Plotting
- Mempelajari cara membuat 3D Plotting dengan Matplotlib

## Import Modules
- Import modules dan menampilkan versi modules

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 3D AXES
- Objek 3D axes perlu membangun sendiri objek figure dan axes
- `plt.figure()` untuk membuat objek figure
- `plt.axes()` untuk membuat objek axes
- `projection='3d'` untuk membentuk axes 3D

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

## 3D Line Plot
- Variabel `z` untuk menghasilkan deret bilangan dari angka 0 sampai 1 sebanyak 100
- Variabel `x` untuk menampung `z * np.sin(20 * z)`
- Variabel `y` untuk menampung `z * np.cos(20 * z)`

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

- `plt.figure()` untuk membuat objek figure
- `plt.axes()` untuk membuat objek axes
- `projection='3d'` untuk membentuk axes 3D

- Pada `ax.plot(x, y, z, color='blue')` parameter 1 sampai 3 untuk menentukan koordinat sedangkan paramaeter ke 4 merupakan opsional dan color ini untuk menentukan warna dari lineplot.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='blue')

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

## Scatter Plot
- Untuk membuat 3D plotting berbentuk Scatter Plot hanya mengganti `ax.scatter(x, y, z, color='yellow')`

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='yellow')

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```
