# 3D Area Plotting
- Mempelajari cara membuat 3D Area Plotting dengan Matplotlib

## Import Modules
- Import modules dan magic command serta menampilkan versi modules

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Membentuk Sample Dataset
- Membentuk suatu deret bilangan yang menghasilkan nilai dari -3 sampai 3 dengan jumlah 100 bilangan
- `np.meshgrid(n, n)` untuk membentuk data surface (data area) yang disimpan dalam variabel x untuk data pertama dan y untuk data kedua
- Variabel Z akan menampung hasil akar kuadrat dari `np.sqrt(x**2 + y**2)`

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## 3D Wireframe
- `plt.figure()` untuk membuat objek figure
- `plt.axes()` untuk membuat objek axes
- `projection='3d'` untuk membentuk axes 3D
- `ax.plot_wireframe(` untuk membentuk plotting dalam bentuk wireframe plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='pink')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

## 3D Surface Plot
- `plt.figure()` untuk membuat objek figure
- `plt.axes()` untuk membuat objek axes
- `projection='3d'` untuk membentuk axes 3D
- `ax.plot_surface()` untuk membentuk plotting dalam bentuk surface plot
- `cmap='viridis'` yaitu color map

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
